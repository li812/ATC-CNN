```python
import numpy as nb
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('dataset/ug_pg.csv')
```

```python


# Count the occurrences of each value in 'UG'
value_counts = df['UG'].value_counts()

# Sort the values by the count and then by the length of the value
sorted_values = value_counts.sort_values(ascending=False).index.tolist()
sorted_values.sort(key=len, reverse=True)

# Get the first 4 longest repeated values
longest_repeated_values = sorted_values[:4]

# Get the counts of the first 4 longest repeated values
counts = [value_counts[value] for value in longest_repeated_values]

# Create a bar plot
plt.figure(figsize=(20, 12))
plt.bar(longest_repeated_values, counts)
plt.xlabel('UG Courses')
plt.ylabel('Counts')
plt.title('First 4 Longest Repeated UG Courses')
plt.show()
```
